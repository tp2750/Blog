```julia
using Catalyst
```

```julia
## Define problem
oct = @reaction_network begin
    k₁,  E + S --> ES
    k₋₁,    ES --> S + E
    k₂,     ES --> P + E
    k₃,  E + P --> EP
    k₋₃,    EP --> E + P
    k₄, E2 + S --> E2S
    k₋₄,    E2S --> E2 + S
end k₁ k₋₁ k₂ k₃ k₋₃ k₄ k₋₄
```

```julia
Graph(oct)
```
